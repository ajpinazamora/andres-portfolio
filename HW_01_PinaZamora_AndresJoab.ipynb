````python
import os
import requests
from zipfile import ZipFile
import tarfile
import pandas as pd
import numpy as np


# Define the URL and file names
zip_url = "https://archive.ics.uci.edu/static/public/117/census+income+kdd.zip"
zip_filename = "census+income+kdd.zip"
tar_gz_filename = "census.tar.gz"
datafile = "census-income.data"

# Create a folder to store the downloaded and extracted files
folder_path = "census"
os.makedirs(folder_path, exist_ok=True)

# Download the zip file
response = requests.get(zip_url, stream=True)
with open(os.path.join(folder_path, zip_filename), "wb") as zip_file:
    for chunk in response.iter_content(chunk_size=128):
        zip_file.write(chunk)

# Extract the tar.gz file
with ZipFile(os.path.join(folder_path, zip_filename), 'r') as zip_ref:
    zip_ref.extract(tar_gz_filename, folder_path)

# Check if the data file already exists and remove it
datafile_path = os.path.join(folder_path, datafile)
if os.path.exists(datafile_path):
    os.remove(datafile_path)

# Extract the tar.gz file
with tarfile.open(os.path.join(folder_path, tar_gz_filename), 'r:gz') as tar_ref:
    tar_ref.extract(datafile, folder_path)

# Load the data into a Pandas DataFrame
df = pd.read_csv(datafile_path, header=None)

# Display the DataFrame
df.head()
````

### Q3 Outlier Identification (25 pts)

#### Q3.1 For each continuous feature, identify the outliers using the IQR method. (15 pts)
For each feature, report the lower and upper bounds and number of instances that are identified as outliers.
Then, display boxplots and discuss if your outliers analysis makes sense. Discuss why you would (or would not) use the IQR method.


````python
outl=[]
for index, row in features_df.iterrows():
    
    if row["scale (N, O, I, R)"] in ["I", "R"]:
        
        col_name = index
        IQR = (df[col_name].quantile(0.75) - df[col_name].quantile(0.25)).round(2)
        Q1 = df[col_name].quantile(0.25).round(2)
        Q3 = df[col_name].quantile(0.75).round(2)
        lower = Q1 - (1.5*IQR)
        upper = Q3 + (1.5*IQR)
        num_outliers = ((df[col_name] < lower) | (df[col_name] > upper)).sum()
        
        outl.append({
            "Feature": col_name,
            "Q1": Q1,
            "Q3": Q3,
            "IQR": IQR,
            "Lower Bound": lower,
            "Upper Bound": upper,
            "Numer of Outliers" : num_outliers
        })
outliers = pd.DataFrame(outl)
outliers
````

````python
for feature in outliers["Feature"]:
    
    sns.boxplot(x=df[feature])
    
    plt.title(f"Boxplot of {feature}")
    plt.show()
````

#### Notes:
In general, the IQR method is not recommended for most of the variables in this dataset because many of them are highly skewed or heavily concentrated at zero. For several financial variables, such as wages, capital gains, capital losses, and dividends, both Q1 and Q3 are equal to zero, which makes the IQR equal to zero and causes all positive values to be incorrectly classified as outliers. These values represent real variation rather than errors. In addition, some variables have natural limits (such as age) where statistical bounds may not align with logical constraints. Because of these characteristics, the IQR method does not always provide meaningful or realistic outlier detection in this dataset and should be applied cautiously. 

#### Q3.2. Replace the outlying values in WKSWORK, MARSUPWT and AAGE features. (10 pts)
Use clamping with upper and lower bounds you found in the previous step. Report how many individual cells are being updated for each feature.


````python
selected = outliers[outliers["Feature"].isin(["WKSWORK", "MARSUPWT", "AAGE"])]
updates = []

for index, row in selected.iterrows():
    
    feature = row["Feature"]
    lower = row["Lower Bound"]
    upper = row["Upper Bound"]
    
    # Count updates BEFORE modifying
    below = (df[feature] < lower).sum()
    above = (df[feature] > upper).sum()
    total_updates = below + above
    
    # Apply clamping manually
    df.loc[df[feature] < lower, feature] = lower
    df.loc[df[feature] > upper, feature] = upper
    
    updates.append({
        "Feature": feature,
        "Cells Updated": total_updates
    })

updates_df = pd.DataFrame(updates)
updates_df
````

### Q4 Normalization (10 pts : Bonus)

Normalize the MARSUPWT, AAGE, NOEMP, CAPGAIN, and CAPLOSS features
* Use range normalization for MARSUPWT feature
* Use robust scaling for NOEMP feature
* Use Z-score normalization for AAGE feature
* Use log scaling for CAPGAIN and CAPLOSS features (impute your zero values with 10 before transformation to avoid undefined  $log_{b}0$) (This is not discussed in class but an important scaling/normalization method)

````python
from sklearn import preprocessing
````

````python
range_norm = df[['MARSUPWT']]
print(range_norm.head())
min_max_scaler = preprocessing.MinMaxScaler()

range_scaled = min_max_scaler.fit_transform(range_norm) 
df_range = pd.DataFrame(data=range_scaled, columns=['MARSUPWT_norm'])
df_range.head()
````

````python
robust_norm = df[['NOEMP']]
print(robust_norm.head())
robust_scaler = preprocessing.RobustScaler()

robust_scaled = robust_scaler.fit_transform(robust_norm) 
df_rob = pd.DataFrame(data=robust_scaled, columns=['NOEMP_norm'])
df_rob.head()
````

````python
z_norm = df[['AAGE']]
print(z_norm.head())
standard_scaler = preprocessing.StandardScaler()

z_scaled = standard_scaler.fit_transform(z_norm) 
df_norm = pd.DataFrame(data=z_scaled, columns=['AAGE_norm'])
df_norm.head()
````
